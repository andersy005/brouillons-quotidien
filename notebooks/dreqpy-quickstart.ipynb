```python
def describe(dq, section):
    print(f'section: {section}')
    for a in dq.coll[section].attDefn.keys():
        print(f"{a}: {dq.coll[section].attDefn[a].title}")
```

```python
for key in sorted(dq.coll.keys()):
    print('_' * 100)
    describe(dq, key)
```

```python
describe(dq, 'CMORvar')
```

```python
def get_items(dq, section, keys, items_range=None):
    if items_range is None:
        items_range = slice(0, 5)
    for r in dq.coll[section].items[items_range]:
        try:
            for key in keys:
                print(f'{key}={r.__dict__[key]}', end=' ')
        except KeyError:
            pass

        print()
```

```python
get_items(dq, 'var', ['label', 'title', 'sn'])
```

```python
get_items(dq, 'CMORvar', ['label', 'title', 'modeling_realm', 'mipTable'])
```

```python
get_items(dq, 'experiment', ['label', 'title', 'mip'])
```

```python
get_items(dq, 'mip', dq.coll['mip'].attDefn.keys())
```

```python
get_items(dq, '__sect__', ['label', 'title', 'labUnique'], slice(0, None))
```
