```python
import numpy as np
import xarray as xr

import dask
import dask.array as da
```

```python
def f(x):
    return 1.1 * x


def g(x):
    return 0.9 * x


num_steps = 1000
num_times = int(1e6)

u = np.ones(num_times)
s = np.ones(num_times)
```

```python
arrays = []
for i in range(num_steps):
    u = dask.delayed(f)(u)
    s = dask.delayed(g)(s)
    product = da.from_delayed(u * s, shape=(num_times,), dtype=float)
    arrays.append(product)
```
