```python
cat_sub = cat.search(
    table_id=['Amon', 'Omon', 'Lmon'],
    member_id=["r1i1p1f1"],
    variable_id=['gpp', 'mrso', 'pr', 'co2', 'thetao', 'no3'],
    grid_label=['gn', 'gr'],
    source_id=['CESM2-WACCM', 'MIROC6', 'CanESM5'],
    experiment_id=['historical', 'ssp370'],
)
cat_sub.df
dsets = cat_sub.to_dataset_dict(storage_options={'anon': True})
```

```python
results = {}

for key, ds in dsets.copy().items():
    subset = {}
    if 'lev' in ds:
        subset['lev'] = 0
    if 'plev' in ds:
        subset['plev'] = 0
    subset['time'] = list(range(6))
    subset['member_id'] = 0

    dset_subset = ds.isel(**subset).load()
    del dset_subset.attrs['intake_esm_varname']
    results[key] = dset_subset
```

```python
results['CMIP/MIROC/MIROC6/historical/Lmon/gn']
```

```python
results['CMIP/MIROC/MIROC6/historical/Omon/gn']
```

```python
tree = datatree.DataTree.from_dict(results)
```
